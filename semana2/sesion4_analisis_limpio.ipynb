```python
#Importa libreria 
# importar dataset Train 
import pandas as pd 
df = pd.read_csv('C:\\Users\\crist\\Proyectos_pyhon\\train.csv')
```

```python
# Informacion de las variables del dataset 
df.info()
```

```python
# Limpieza datos vacios o nulos 
# Se rellena Age(edad) con la Media, Para no tener valores extremos como con la mediana o la moda
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Age'].isnull().sum()
```

```python
# La Variable Cabin Al tener la el (68%) de sus datos vacios no se realiza relleno de datos
# en alguna posible decision que se deba tomar, lo hariamos con la variable Pclass 
```

```python
# La variable Embarked se rellena con la moda al tener solo el (0,02) de los datos vacios
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
print(df['Embarked'].isnull().sum())
```

```python
#Verificacion de las variables no se encuentren con valores vacios
#Variable (Cabin) queda con el 204
df.info()
```
