```python
# Importar librerias para limipieza y visualizacion 
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\crist\\Proyectos_pyhon\\train.csv')
df.info()
```

```python
# Realizamos limpieza de datos para evitar datos vacios 
# Age utilizamos la Mediana para evitar valores extremos como con la media o la moda
# Embraked Utilizamos la moda por que solo vamos a sustituir 2 vacios dentro de la variable
# Cabin Es una variable con 63% de los datos vacios, (No se realiza cambios) para posible analisis utilizamos Pclass con Variable alternativa

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked']= df['Embarked'].fillna(df['Embarked'].mode()[0])
df.info()
```

```python
conteo = df['Survived'].value_counts()
print(conteo)
```

```python
categorias = ['No Sobrevivio' , 'Sobrevivio']
valores = [549, 342]
plt.figure(figsize=(9, 5))
plt.bar(categorias, valores)
plt.title('Supervivencia de pasajeros — Titanic')
plt.show()
```

```python
categorias = ['No Sobrevivió', 'Sobrevivió']
valores = [549, 342]
colores = ['#ef4444', "#258666"]

plt.figure(figsize=(9, 5))
plt.bar(categorias, valores, color=colores)
plt.title('Supervivencia de pasajeros — Titanic')
plt.ylabel('Tasa de sobrevivencia (%)')

for i, v in enumerate(valores):
    plt.text(i, v + 1, f'{v}%', ha='center', fontsize=12 fontweight= 'bold')
plt.tight_layout
plt.show()
```
