```python
import pandas as pd
df = pd.read_csv ('C:\\Users\\crist\\Proyectos_pyhon\\train.csv')
```

```python
# Se revisa variables con valores Vacios para limpiar el Dataset y no tomar decisiones con sesgadas  
df.isnull().sum()
```

```python
# Age: se rellena con la mediana (28) en lugar de la media
# porque hay valores extremos (bebés y ancianos) que distorsionan la media.
# La mediana representa mejor al pasajero típico.
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Verificación  — resumen de nulos tras limpieza
df['Age'].isnull().sum()
```

```python
# Cabin: se mantiene con nulos (77% de valores vacíos).
# No es fiable para análisis. Se usará Pclass como alternativa
# para representar el nivel socioeconómico del pasajero.
```

```python
# Embarked: se usa la moda para rellenar los 2 valores nulos.
# El porcentaje de nulos (0.2%) es muy bajo para usar media o mediana.
# La moda es 'S' (Southampton) — el puerto más frecuente.
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked'].isnull().sum()
```

```python
# Verificación  final — resumen de nulos tras limpieza
df.isnull().sum()
```
